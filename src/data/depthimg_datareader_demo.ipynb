```python
import re
import numpy as np
```

```python
def read_pfm(path):
    with open(path, 'rb') as f:
        header = f.readline().decode('ascii').rstrip()
        if header not in ('PF, Pf'):
            raise Exception('Not a PFM file.')

        dims = f.readline().decode('ascii').strip()
        while dims.startswith('#'):
            dims = f.readline().decode('ascii').strip()
        m = re.match(r'^(\d+)\s+(\d+)$', dims)
        if not m:
            raise ValueError('Malformed PFM header(width height).')
        width, height = map(int, m.groups())
        scale = float(f.readline().decode('ascii').strip())
        endian = '<' if scale < 0 else '>'
        data = np.fromfile(f, endian + 'f')
        channels = 3 if header == 'PF' else 1
        expected = width * height * channels
        if data.size != expected:
            raise ValueError('Malformed PFM. Expected %d floats, got %d.')

        data = np.reshape(data, (height, width,
                                 channels)) if channels == 3 else np.reshape(
                                     data, (height, width))
        data = np.flipud(data)
        return data, abs(scale)
```

```python
pfm_path = '/home/LionelZ/Pythoncode/EDOF-Meta-bifocal/src/data/0000000.pfm'
arr, scale = read_pfm(pfm_path)
print(f'PFM shape: {arr.shape}, scale: {scale}')
```
